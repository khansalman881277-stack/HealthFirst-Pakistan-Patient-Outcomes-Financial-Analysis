```python
# Convert column to string type (so text operations like replace can work safely)
df['bill_pkr'] = df['bill_pkr'].astype(str)

# Remove 'Rs.' prefix from values (e.g., "Rs.25000" → "25000")
df['bill_pkr'] = df['bill_pkr'].str.replace('Rs.', '', regex=False)

# Remove commas from numbers (e.g., "25,000" → "25000")
df['bill_pkr'] = df['bill_pkr'].str.replace(',', '', regex=False)

# Remove any extra spaces before or after values
df['bill_pkr'] = df['bill_pkr'].str.strip()

# Convert cleaned text into numeric format
# If any value cannot be converted, it becomes NaN (missing value)
df['bill_pkr'] = pd.to_numeric(df['bill_pkr'], errors='coerce')
```

Question no 01

Which 5 diagnoses generate the highest average bill?

```python
import matplotlib.pyplot as plt

# Group, calculate stats, and get top 5 in one go
top5 = (
    df.groupby('diagnosis')['bill_pkr']
      .agg(['mean', 'median', 'count'])
      .round(2)
      .sort_values(by='mean', ascending=False)
      .head(5)
)

# Display
print("\nTop 5 Diagnoses by Average Bill:\n")
print(top5)

# Plot
plt.bar(top5.index, top5['mean'])
plt.xticks(rotation=45)
plt.title('Top 5 Diagnoses by Average Bill')
plt.tight_layout()
plt.show()
```

Question no 02

Do insured patients stay longer in hospital than
uninsured?

```python
from scipy.stats import ttest_ind, mannwhitneyu

# Split data into two groups
insured = df[df['insurance'] == 'Yes']['length_of_stay_days']
uninsured = df[df['insurance'] == 'No']['length_of_stay_days']

# Mean & Median
print("INSURED -> Mean:", insured.mean(), "Median:", insured.median())
print("UNINSURED -> Mean:", uninsured.mean(), "Median:", uninsured.median())

# T-test
t_stat, p_value_t = ttest_ind(insured, uninsured, nan_policy='omit')

# Mann-Whitney test
u_stat, p_value_u = mannwhitneyu(insured, uninsured)

print("\nT-test p-value:", p_value_t)
print("Mann-Whitney p-value:", p_value_u)
```

```python
crosstab = pd.crosstab(df['ward'], df['outcome'])
print(crosstab)
```

```python
crosstab_pct = crosstab.div(crosstab.sum(axis=1), axis=0) * 100
print(crosstab_pct.round(2))
```

```python
crosstab_pct.plot(kind='bar', stacked=True)

plt.title('Patient Outcome by Ward Type')
plt.xlabel('Ward')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.legend(title='Outcome')
plt.tight_layout()
plt.show()
```
